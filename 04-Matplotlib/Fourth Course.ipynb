```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.arange(0, 10)
```

```python
y = x**2
```

```python
plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(0, 9.5)
plt.ylim(0, 90)
plt.title("My Title")
plt.figure(figsize=(10,10))
plt.savefig("myfirstplot.png");
```

---

```python
a = np.linspace(0, 10, 11)
```

```python
b = a**3
```

```python
fig = plt.figure()

# Large plot
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
axes.set_xlim(0, 10.5)
axes.set_ylim(0, 1050)
axes.set_xlabel("Large Plot's X axis")
axes.set_ylabel("Large Plot's Y axis")
axes.set_title("Large Plot (** of 4)")

# Small plot
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])
axes2.plot(x, y)
axes2.set_xlim(0, 10)
axes2.set_ylim(0, 100)
axes2.set_xlabel("Small Plot's X axis")
axes2.set_ylabel("Small Plot's Y axis")
axes2.set_title("Small Plot (** of 2)")
```

---

```python
fig = plt.figure(figsize=(4, 3), dpi=150)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(a, b)

fig.savefig("new_fig_2", bbox_inches="tight")
```

---

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)

plt.tight_layout()
#fig.suptitle("Figure Level", fontsize=16)
#fig.subplots_adjust(wspace=0.9, hspace=0.9)
```

---

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, label="x vs y")
axes.plot(x, y/2, label="x vs y/2")

axes.legend(loc=(1.05, 0.905))
```
