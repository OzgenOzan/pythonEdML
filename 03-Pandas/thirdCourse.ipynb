```python
def tipper(total_bill, tip):
    if tip/total_bill >= 0.25:
        return "nice tipper"
    elif 0.1 <= tip/total_bill < 0.25:
        return "so so"
    else:
        return "tip's dead"
```

```python
dfTest[["total_bill", "tip"]].apply(lambda arg: tipper(arg["total_bill"], arg["tip"]), axis=1)
```

```python
dfTest["tipper"] = np.vectorize(tipper)(dfTest["total_bill"], dfTest["tip"])
```
