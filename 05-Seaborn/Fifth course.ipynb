```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")

df.head()
```

```python
pearsonr(df["salary"], df["sales"]) # to calculate correlation
```

```python
plt.figure(figsize=(12, 6), dpi=300)

sns.scatterplot(x="salary", y="sales", data=df, hue="level of education", size="work experience", alpha=0.80)

plt.savefig("test1_fig")
```

```python
df = pd.read_csv("StudentsPerformance.csv")

df.head()
```

```python
plt.figure(figsize=(12, 8), dpi=300)

sns.boxplot(data=df, x="test preparation course", y="math score")
```

```python
df["avg score"] = round((df["math score"]+df["reading score"]+df["writing score"])/3, ndigits=2)
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 8), dpi=300)

sns.boxplot(data=df, x="test preparation course", y="avg score");
```

```python
plt.figure(figsize=(12, 8), dpi=300)

sns.boxplot(data=df, x="parental level of education", y="avg score", hue="test preparation course", palette="Set2")

plt.legend(bbox_to_anchor=(1.15, 1.013));
```

```python
plt.figure(figsize=(12, 8), dpi=200)

sns.violinplot(data=df, x="parental level of education", y="avg score");
```

---

```python
sns.jointplot(data=df, x="math score", y="avg score", kind="hex");
```

```python
sns.pairplot(data=df[["math score", "reading score", "writing score", "gender"]], hue="gender", corner=True);
```
